# Aggregating data with transform ()

```python
# Create a new loyalty flag using loc
df_ords_prods_merged['max_order'] = df_ords_prods_merged.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Check
df_ords_prods_merged.head(15)
```

# Deriving columns with loc

```python
# create a loyalty flag 
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['max_order'] <= 40) & (df_ords_prods_merged['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Check 
df_ords_prods_merged['loyalty_flag'].value_counts(dropna = False)
```

```python
df_ords_prods_merged[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

```python
# Perform aggregation to check basic statistics
df_ords_prods_merged.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

Question 5

New customers spend more, on average, than loyal customers and regular customers. Loyal customers spend less on average.

```python
# Create a spending flag for each user based on the average price across all their orders
df_ords_prods_merged['mean_spend'] = df_ords_prods_merged.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_ords_prods_merged.head(15)
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['mean_spend'] < 10, 'spender_flag'] = 'Low spender'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['mean_spend'] >= 10, 'spender_flag'] = 'High spender'
```

```python
# Check flag
df_ords_prods_merged['spender_flag'].value_counts(dropna = False)
```

```python
df_ords_prods_merged.head()
```

```python
# Create an order frequency flag that marks the regularity of a userâ€™s ordering behavior according to the median in the â€œdays_since_prior_orderâ€ column
df_ords_prods_merged['median_order_frequency'] = df_ords_prods_merged.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_ords_prods_merged.head(15)
```

```python
df_ords_prods_merged['median_order_frequency'].isnull().sum()
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['median_order_frequency'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
```

```python
df_ords_prods_merged.loc[(df_ords_prods_merged['median_order_frequency'] <= 20) & (df_ords_prods_merged['median_order_frequency'] > 10), 'order_frequency_flag'] = 'Regular customer'
```

```python
df_ords_prods_merged.loc[df_ords_prods_merged['median_order_frequency'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
#Check flag
df_ords_prods_merged['order_frequency_flag'].value_counts(dropna = False)
```
